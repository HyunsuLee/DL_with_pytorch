```python
%matplotlib inline
import numpy as np
import torch
torch.set_printoptions(edgeitems=2, linewidth=75)
```

```python
t_c = [0.5,  14.0, 15.0, 28.0, 11.0,  8.0,  3.0, -4.0,  6.0, 13.0, 21.0] # measure from new temp sensor
t_u = [35.7, 55.9, 58.2, 81.9, 56.3, 48.9, 33.9, 21.8, 48.4, 60.4, 68.4] # measured in celcius
t_c = torch.tensor(t_c)
t_u = torch.tensor(t_u)
```

```python
def model(t_u, w, b):
    return w * t_u + b # model's output is temperature in celcius
```

```python
def loss_fn(t_p, t_c):
    squared_diffs = (t_p - t_c)**2
    return squared_diffs.mean() # MSE mean square less
```

```python
w = torch.ones(())
b = torch.zeros(())

t_p = model(t_u, w, b)
t_p
```

```python
loss = loss_fn(t_p, t_c)
loss
```

```python
# example code for understang broadcasting
x = torch.ones(())
y = torch.ones(3,1)
z = torch.ones(1,3)
a = torch.ones(2, 1,1,1)
print(f"shapes: x: {x.shape}, y: {y.shape}")
print(f"        z: {z.shape}, a: {a.shape}")
print("x * y:", (x * y).shape)
print("y * z:", (y * z).shape)
print("y * z * a:", (y * z * a).shape)
print("x * a:", (x * a).shape)
print("y * a:", (y * a).shape)
print("z * a:", (z * a).shape)
print("x * y * a:", (x * y * a).shape)
```

* x * y는 1. * 1,1,1 => 1,1,1 broadcasting
* y * z 는 3,1 tensor와 1,3 tensor가 만나 3,3 tensor로 broadcasting된것
* 길이가 1인 차원이 늘어나면서 계산된다.
* 만약 차원의 사이즈가 안 맞고, 원소가 안 맞으면, 한쪽에 길이가 1인 tensor가 있어야 한다. 
* 만약 한쪽에 길이악 1인 tensor가 맞춰야 하는 tensor의 차원보다 많은 경우, broadcasting은 뒤쪽 차원부터 된다.
* 즉 남는 singleton tensor는 앞 쪽 axis에 배치된다.



```python
delta = 0.1

loss_rate_of_change_w = \
    (loss_fn(model(t_u, w + delta, b), t_c) - 
     loss_fn(model(t_u, w - delta, b), t_c)) / (2.0 * delta)
# calculate derivative upon to weight, delta = \delta h in calculus
```

```python
learning_rate = 1e-2

w = w - learning_rate * loss_rate_of_change_w # new weight
```

```python
loss_rate_of_change_b = \
    (loss_fn(model(t_u, w, b + delta), t_c) - 
     loss_fn(model(t_u, w, b - delta), t_c)) / (2.0 * delta)

b = b - learning_rate * loss_rate_of_change_b
```

```python
def dloss_fn(t_p, t_c):
    dsq_diffs = 2 * (t_p - t_c) / t_p.size(0)  # t_p : model output, t_c : temp in celcius (ground truth)
    return dsq_diffs # this tensor has same dimension with t_p, thus it is vector gradient according to each model's output
```

```python
dloss_fn(t_p, t_c).size(0)
```

```python
def dmodel_dw(t_u, w, b): # model t_p = t_u * w + b > derivative to w
    return t_u
```

```python
def dmodel_db(t_u, w, b): # derivative to b
    return 1.0
```

```python
def grad_fn(t_u, t_c, t_p, w, b):
    dloss_dtp = dloss_fn(t_p, t_c)
    dloss_dw = dloss_dtp * dmodel_dw(t_u, w, b) # element wise multiplication
    dloss_db = dloss_dtp * dmodel_db(t_u, w, b)
    return torch.stack([dloss_dw.sum(), dloss_db.sum()])  
```

```python
grad_fn(t_u, t_c, t_p, w, b)
```

```python
def training_loop(n_epochs, learning_rate, params, t_u, t_c):
    for epoch in range(1, n_epochs + 1):
        w, b = params

        t_p = model(t_u, w, b)  # <1>
        loss = loss_fn(t_p, t_c)
        grad = grad_fn(t_u, t_c, t_p, w, b)  # <2>

        params = params - learning_rate * grad

        print('Epoch %d, Loss %f' % (epoch, float(loss))) # <3>
            
    return params
```

```python
def training_loop(n_epochs, learning_rate, params, t_u, t_c,
                  print_params=True):
    for epoch in range(1, n_epochs + 1):
        w, b = params

        t_p = model(t_u, w, b)  # <1>
        loss = loss_fn(t_p, t_c)
        grad = grad_fn(t_u, t_c, t_p, w, b)  # <2>

        params = params - learning_rate * grad

        if epoch in {1, 2, 3, 10, 11, 99, 100, 4000, 5000}:  # <3>
            print('Epoch %d, Loss %f' % (epoch, float(loss)))
            if print_params:
                print('    Params:', params)
                print('    Grad:  ', grad)
        if epoch in {4, 12, 101}:
            print('...')

        if not torch.isfinite(loss).all():
            break  # <3>
            
    return params
```

```python
training_loop(
    n_epochs = 100, 
    learning_rate = 1e-2, 
    params = torch.tensor([1.0, 0.0]), 
    t_u = t_u, 
    t_c = t_c)
```

```python
training_loop(
    n_epochs = 100, 
    learning_rate = 1e-4, 
    params = torch.tensor([1.0, 0.0]), 
    t_u = t_u, 
    t_c = t_c)
```

```python
t_un = 0.1 * t_u # scaling the input data. 
t_un
```
